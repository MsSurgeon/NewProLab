```python
def url2domain(url):
    try:
        a = urlparse(unquote(url.strip()))
        if (a.scheme in ['http','https']):
            if a.netloc == 'news.yandex.ru':
                if(re.search('(?<=cl4url=).+(html)', url)):
                    url = 'http://' + re.search('(?<=cl4url=).+(html)', url).group(0)
                    a = urlparse(unquote(url.strip()))
                    b = re.search("(?:www\.)?(.*)",a.netloc).group(0)
                else:
                    url = 'http://' + re.search('(?<=cl4url=).+', url).group(0)
                    a = urlparse(unquote(url.strip()))
                    b = re.search("(?:www\.)?(.*)",a.netloc).group(0)
            else:
                b = re.search("(?:www\.)?(.*)",a.netloc).group(1)
            if b is not None:
                stripUrl = str(b).strip()
                return stripUrl
                if stripUrl in url2MetaUrl:
                    if url2MetaUrl[stripUrl]:
                        return url2MetaUrl[str(b).strip()]
                    else:
                        return ''
                else:
                    return stripUrl
            else:
                return ''
        else:
            return ''
    except:
        return ''
```

```python
curUDict = {}

for l in open('gender_age_dataset_h.txt'):
    parts = l.strip().split('\t')
    if len(parts) < 4:
        continue
    try:
        localDict = {}
        j = json.loads(parts[3])
        for u in j['visits']:
            curUrl = url2domain(u['url'])
            if curUrl and not curUrl in localDict:
                localDict[curUrl] = 1
                if curUrl in curUDict:
                    curUDict[curUrl] = curUDict[curUrl] + 1
                else:
                    curUDict[curUrl] = 1
    except:
        continue
```

```python
urls = [x for x in sorted(curNewDict)]

curLen = len(urls)
uniqueTopic = {}

for u in topicDict:
    uniqueTopic[topicDict[u]] = topicDict[u]

for u in uniqueTopic:
    urls.append(u)
    uniqueTopic[u] = curLen
    curLen += 1

checkFields = ['gender', 'age1', 'age2', 'age3', 'age4', 'age5']
len(urls)
```

```python
knownSet = []
checkSet = []
unknownSet = []
uidSet = []

urlLen = len(urls)

age2idx = {
    '18-24': 1,
    '25-34': 2,
    '35-44': 3,
    '45-54': 4,
    '>=55': 5,
}

excludeTopics = {u'Реклама': True, u'Поисковые системы': True, u'Баннерные сети': True}

for l in open('gender_age_dataset_h.txt'):
    userSet = [0] * urlLen
    parts = l.strip().split('\t')
    added = False
    if len(parts) < 4:
        print('skip ' + l.strip())
        continue
    try:
        j = json.loads(parts[3])
        for u in j['visits']:
            curUrl = url2domain(u['url'])
            if curUrl in curNewDict:
                if curUrl in topicDict:
                    if not topicDict[curUrl] in excludeTopics:
                        added = True
                        if not userSet[curNewDict[curUrl]]:
                            userSet[uniqueTopic[topicDict[curUrl]]] += 1
                        userSet[curNewDict[curUrl]] = 1
                else:
                    added = True
                    userSet[curNewDict[curUrl]] = 1
    except Exception as e:
        print('except!! ' + str(e))
        continue
        
    if parts[0] == '-' or parts[1] == '-': 
        uidSet.append([added])        
        unknownSet.append(userSet)         
    else:        
        if not added:
            continue   
        
        knownSet.append(userSet)
        
        curCheckSet = [0] * 6
        if parts[0] == 'F':            
            curCheckSet[0] = 0
        else:  
            curCheckSet[0] = 1
        curCheckSet[age2idx[parts[1]]] = 1
        checkSet.append(curCheckSet)

len(knownSet)
```

```python
emptyColumns = []
for j in range(len(urls)):
    ifKnown = False
    ifUnknown = False
    if True in any(i > 0 for i in [row[j] for row in knownSet]):
        continue
    else:
        ifKnown = True   
    if True in any(i > 0 for i in [row[j] for row in unknownSet]):
        continue
    else:
        ifUnknown = True
    if ifKnown and ifUnknown:
        emptyColumns.append(j)
        
len(emptyColumns)
```
