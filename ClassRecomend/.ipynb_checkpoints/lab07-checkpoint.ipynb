```python
import pandas as pd
import numpy as np
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from matplotlib import pyplot as plt
%matplotlib inline
from pylab import rcParams
import warnings
warnings.simplefilter('ignore')
```

```python
from pymystem3 import Mystem
m = Mystem()
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(u'[A-Za-zА-Яа-яёЁ]+')
```

```python
def cos_m(x, y):
    return abs(cosine_similarity(x, y)[0][0])
```

```python
def tokenize(text):
    tokens = tokenizer.tokenize(text)
    lemms = []
    for i in tokens:
        if len(i)>2:
            lemms.append(m.lemmatize(i)[0])
    return lemms
```

```python
df_es = pd.DataFrame()
df_es = df[df['lang'] == 'es']
df_es.reset_index(inplace = True, drop = True)

es_desc = list(df_es['desc'])
tf_v_es = TfidfVectorizer(norm=None, smooth_idf=False)
matrix_es = tf_v_es.fit_transform(es_desc).toarray()

cos_5504 = []
cos_8944 = []
for i in range(matrix_es.shape[0]):
    cos_5504.append(cos_m(matrix_es[df_es[df_es['id'] == 5504].index[0]], matrix_es[i]))
    cos_8944.append(cos_m(matrix_es[df_es[df_es['id'] == 8944].index[0]], matrix_es[i]))

df_es['cos_5504'] = cos_5504
df_es['cos_8944'] = cos_8944

df_es.sort_values(by = ['cos_5504', 'name', 'id'], ascending=[False, True, True], inplace = True)
top10_5504 = list(df_es['id'].iloc[1:11])
df_es.sort_values(by = ['cos_8944', 'name', 'id'], ascending=[False, True, True], inplace = True)
top10_8944 = list(df_es['id'].iloc[1:11])
```

```python
df_en = pd.DataFrame()
df_en = df[df['lang'] == 'en']
df_en.reset_index(inplace = True, drop = True)

en_desc = list(df_en['desc'])
tf_v_en = TfidfVectorizer(norm=None, smooth_idf=False)
matrix_en = tf_v_en.fit_transform(en_desc).toarray()

cos_8553 = []
cos_28034 = []
for i in range(matrix_en.shape[0]):
    cos_8553.append(cos_m(matrix_en[df_en[df_en['id'] == 8553].index[0]], matrix_en[i]))
    cos_28034.append(cos_m(matrix_en[df_en[df_en['id'] == 28034].index[0]], matrix_en[i]))

df_en['cos_8553'] = cos_8553
df_en['cos_28034'] = cos_28034

df_en.sort_values(by = ['cos_8553', 'name', 'id'], ascending=[False, True, True], inplace = True)
top10_8553 = list(df_en['id'].iloc[1:11])
df_en.sort_values(by = ['cos_28034', 'name', 'id'], ascending=[False, True, True], inplace = True)
top10_28034 = list(df_en['id'].iloc[1:11])
```
