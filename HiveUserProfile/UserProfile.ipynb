```python
def url2domain(row):
    try:
        a = urlparse(unquote(row['url'].strip()))
        if (a.scheme in ['http','https']):
            b = re.search("(?:www\.)?(.*)",a.netloc).group(1)
            if b is not None:
                return str(b).strip()
            else:
                return ''
        else:
            return ''
    except:
         return

df['dom'] = df.apply(url2domain,axis=1)
```

```python
df = df.drop('url',1)
df = df.drop('ts',1)
```

```python
def if_autouser(row):
    if row['uid'] in autouser:
        return 1
    else:
        return 0

dfs['autouser']  = dfs.apply(if_autouser, axis=1)
dfs
```

```python
from __future__ import division

getcontext().prec = 20

num_row = Decimal(len(dfs))
num_row_cliked = Decimal(len(dfs[dfs.autouser==1]))

domauto = []

for dom in tqdm(dfs.dom.unique()):
    num_dom_clicked = Decimal(len(dfs[(dfs.dom == dom) & (dfs.autouser == 1)]))
    num_all_dom = Decimal(len(dfs[dfs.dom == dom]))

    relevance = pow((num_dom_clicked/num_row),2)/((num_all_dom/num_row)*(num_row_cliked/num_row))
    domauto.append([dom,"{:.20f}".format(relevance)])
df_domauto = pd.DataFrame(domauto,columns=['domain','relevance'])
df_domauto
```

```python
df_relevance = df_domauto[df_domauto.relevance>0].sort_values(by=['relevance', 'domain'], ascending=[False,True]).head(200)

df_domauto.sort_values(by=['relevance', 'domain'], ascending=[False,True]).head(200)
```
